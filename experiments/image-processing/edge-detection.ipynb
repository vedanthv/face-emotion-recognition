```python
import cv2
import os
from matplotlib import pyplot as plt
image = '/kaggle/input/fer2013/train/happy/Training_95673040.jpg'
img = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(img,100,200)
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#function
inp = '/kaggle/input/fer2013/train'
out = '/kaggle/working/edge_detected_images'
os.makedirs(out, exist_ok=True)

# Loop through images and apply edge detection
for filename in os.listdir(inp):
    if filename.endswith('.jpg'):
        image_path = os.path.join(inp, filename)
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        edges = cv2.Canny(img, threshold1=100, threshold2=200)

        output_path = os.path.join(out, f'edge_{filename}')
        cv2.imwrite(output_path, edges)
```
