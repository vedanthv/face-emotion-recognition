```python
#median denoising
import os
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Set up paths
train_dir = "/content/drive/MyDrive/demo"
output_dir = "output"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Emotion category folder names
emotion_names = ["angry", "disgust", "fear", "happy", "sad", "surprised", "neutral"]

# Loop through each emotion category
for emotion, emotion_name in enumerate(emotion_names):
    emotion_dir = os.path.join(train_dir, emotion_name)
    for image_file in os.listdir(emotion_dir):
        image_path = os.path.join(emotion_dir, image_file)

        # Load the image
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply median denoising
        denoised_image = cv2.medianBlur(image, 3)  # Adjust kernel size as needed

        # Save the denoised image in the corresponding output folder
        output_emotion_dir = os.path.join(output_dir, str(emotion))
        os.makedirs(output_emotion_dir, exist_ok=True)
        output_image_path = os.path.join(output_emotion_dir, image_file)
        cv2.imwrite(output_image_path, denoised_image)

        # Display the original and denoised images using Matplotlib
        plt.figure(figsize=(8, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(image, cmap="gray")
        plt.title("Original Image")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(denoised_image, cmap="gray")
        plt.title("Denoised Image")
        plt.axis("off")

        plt.show()
```

```python
#gaussian denoising
import os
import cv2
import matplotlib.pyplot as plt
from google.colab import files

# Set up paths
train_dir = "/content/drive/MyDrive/demo"
output_dir = "output"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Emotion category folder names
emotion_names = ["angry", "disgust", "fear", "happy", "sad", "surprised", "neutral"]

# Loop through each emotion category
for emotion, emotion_name in enumerate(emotion_names):
    emotion_dir = os.path.join(train_dir, emotion_name)
    for image_file in os.listdir(emotion_dir):
        image_path = os.path.join(emotion_dir, image_file)

        # Load the image
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply Gaussian denoising
        denoised_image = cv2.GaussianBlur(image, (5, 5), 0)  # Adjust kernel size and sigma

        # Save the denoised image in the corresponding output folder
        output_emotion_dir = os.path.join(output_dir, str(emotion))
        os.makedirs(output_emotion_dir, exist_ok=True)
        output_image_path = os.path.join(output_emotion_dir, image_file)
        cv2.imwrite(output_image_path, denoised_image)

        # Display the original and denoised images using Matplotlib
        plt.figure(figsize=(8, 4))  # Adjust the figsize as needed
        plt.subplot(1, 2, 1)
        plt.imshow(image, cmap="gray")
        plt.title("Original Image")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(denoised_image, cmap="gray")
        plt.title("Denoised Image")
        plt.axis("off")

        plt.show()
```
